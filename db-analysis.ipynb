```python
df = pd.read_csv('/kaggle/input/vacancies/vacancies_last.csv', sep=";")
df
```

```python
df[['GENERAL_NAME']] = df[['GENERAL_NAME']].apply(lambda x: x.str.strip())
#pd.set_option('display.max_columns', None)  # or 1000
#pd.set_option('display.max_rows', None)  # or 1000
```

Я предполагаю, что "ученый по данным" и "аналитик по данным" об одном и том же

```python
df = df.replace('Ученый по данным', 'Аналитик по данным')
```

**Также проверим, в каким компаниях и городах больше всего вакансий, какой график и какие навыки требуются**

```python
df_science = data_science.groupby(['ORGANIZATION']).count()
df_science.sort_values(by=['VACANCY_NAME'])
```

```python
df_science = data_science.groupby(['WORK_SCHEDULE']).count()
df_science.sort_values(by=['VACANCY_NAME'])
```

```python
new_skills = []
new_city = []
for index, row in data_science.iterrows():
    new_skills.append(row['REQUIREMENTS'].split(' , '))
    new_city.append(row['CITY'].split(', '))
    
data_science['skills'] = new_skills
data_science['city'] = new_city

mlb = MultiLabelBinarizer()
mlb.fit_transform(data_science['skills'])
skills_transform = mlb.transform(data_science['skills'])
classes = list(mlb.classes_)
```

```python
mlb_1 = MultiLabelBinarizer()
mlb_1.fit_transform(data_science['city'])
city_transform = mlb_1.transform(data_science['city'])
city_classes = list(mlb_1.classes_)
```

```python
data_science[classes] = skills_transform
data_science[city_classes] = city_transform
```
